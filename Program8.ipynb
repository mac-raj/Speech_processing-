#  Write Python program for Mel-frequency cepstral coefficients (MFCC)

```python
import librosa 
```

```python
# Load audio file
audio,sr= librosa.load('/content/harvard.wav')

# Compute MFCC features
mfcc = librosa.feature.mfcc(y=audio,sr=sr,n_mfcc=4)
```

```python
# Display MFCC features 
mfcc
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
librosa.display.specshow(mfcc,x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()
```
