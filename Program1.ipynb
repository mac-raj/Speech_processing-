```python
import librosa 
import librosa.display
import matplotlib.pyplot as plt
```

```python
# Define visualize audio signal

def visualize_audio_signal(filename):
  # Load audio signal using librosa 
  y,sr = librosa.load(filename)

  # Compute the short time Fourier transform(STFT) of the signal
  stft = librosa.stft(y)

  # Convert the STFT to decibles (db) using the log-magnetude spectogram
  log_mag_spec = librosa.amplitude_to_db(abs(stft))

  # Plot the spectogram using librosa display specshow
  librosa.display.specshow(log_mag_spec, sr=sr,y_axis='linear')

  # Add a colorbar to the plot
  plt.colorbar(format='%1.0f db')

  # Set the title of the plot 
  plt.title('Spectogram of ' + filename)

  # Shoe the plot 
  plt.show()

filename = '/content/gdrive/MyDrive/Audio_signal_data/sample.wav'
visualize_audio_signal(filename)
```
